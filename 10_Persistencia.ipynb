1. Escribir un programa que llene un archivo .txt con el conteo de 1 hasta un valor que ingrese el usuario.

```python
numero = int(input('Ingrese número: '))
f_conteo = open('conteo.txt', 'w')
for i in range(1, numero+1):
  f_conteo.write(str(i) + '\n')
f_conteo.close()
```

2. Escribir un programa que vaya llenado un diccionario vacío con información sobre una persona (nombre, edad, sexo, teléfono, correo electrónico, etc.) la cual se va ingresando a medida que el programa le pregunte. Cuando termine de ejecutar, se debe almacenar en un archivo llamado 'datos.txt' de manera tabulada.

```python
dict_1 = {}
opcion = input('Desea ingresar dato? (s/n): ')
while opcion == 's':
  clave = input('Ingrese categoría: ')
  valor = input('Ingrese su valor: ')
  dict_1[clave] = valor
  opcion = input('Desea ingresar dato? (s/n): ')
print(dict_1)
```

```python
f_datos = open('datos.txt', 'w')
for j in dict_1:
  f_datos.write(j + '\t' + dict_1[j] + '\n')
f_datos.close()
```
